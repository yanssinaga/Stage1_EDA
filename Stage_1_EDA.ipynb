# UNIVARIATE ANALYSIS

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
Categoric = ['Geography','Gender','NumOfProducts', 'HasCrCard','IsActiveMember', 'Exited']
Numeric = ['CreditScore', 'Age', 'Tenure', 'Balance', 'EstimatedSalary']
```

```python
for i in range(0, len(Numeric)):
  plt.subplot(1, len(Numeric), i+1)
  sns.boxplot(y=data[Numeric[i]], color='skyblue', orient='v')
  plt.tight_layout()
```

## Interpretasi
Berdasarkan visualisasi scatter plot pada beberapa kolom terdapat 2 kolom yang memiliki outlier yaitu pada kolom 'Age' dan 'credit score'.

```python
plt.figure(figsize=(10, 15))
for i in range(0, len(Numeric)):
  plt.subplot(len(Numeric), 3, i+1)
  sns.kdeplot(x=data[Numeric[i]], color='skyblue')
  plt.xlabel(Numeric[i])
  plt.tight_layout()
```

## Interpretasi
- Distibusi Plot Credit Score merupakan distribusi normal.
- Distribusi Plot Age merupakan memiliki distribusi Positively Skewed.
- Distribusi Plot kolom Tenure berdistribusi normal
- Distribusi Plot kolom Balance berdistribusi bimodal
- Distribusi Plot kolom EstimatedSalary berdistribusi normal

```python
plt.figure(figsize=(7, 15))
for i in range (0, len(Categoric)):
  plt.subplot(len(Categoric), 2, i+1)
  sns.countplot(x=data[Categoric[i]], color='salmon')
  plt.tight_layout()
```

Berdasarkan analisis yang dilakukan diperoleh bahwa data target (Exited -> Churn or No) merupakan data yang imbalance karena data "0" itu jauh lebih banyak dibandingkan data "1".<br>
Selain itu, pada kolom Geography, NumOfProducts dan HasCrCard dapat dilihat bahwa terjadi ketimpangan data.

# BUSINESS INSIGHT

1. Credit Score dan Churn Rate: Korelasi antara credit score dengan churn rate menunjukkan di mana pelanggan dengan credit score rendah cenderung berpotensi churn, dan sebaliknya.
Rekomendasi:
Berdasarkan insight tersebut, bank dapat mengidentifikasi cara-cara yang mendorong nasabah dengan skor kredit yang lebih rendah untuk tetap berada di bank (membuat nasabah tidak churn).

2.  Pengaruh Keaktifan Member terhadap Churn Rate :
Dari insight yang ditemukan, terlihat bahwa pelanggan yang tidak aktif cenderung memiliki tingkat churn yang lebih tinggi.
Rekomendasi :
berdasarkan insight, penting bagi bisnis untuk mendorong pelanggan aktif melalui penawaran unik, atau pelayanan yang lebih baik, sehingga dapat mengurangi churn dan meningkatkan retensi pelanggan.

3. Distribusi Umur Pelanggan : menunjukkan korelasi antara usia pelanggan dengan keputusan untuk churn.
Rekomendasi:
Berdasarkan insight ini, bisnis dapat memahami bagaimana faktor usia mempengaruhi keputusan pelanggan untuk tetap berlangganan atau churn. Bisnis dapat mengembangkan strategi retensi yang khusus untuk kelompok usia tertentu, seperti memberikan promosi yang relevan atau penawaran yang disesuaikan dengan kebutuhan pelanggan di berbagai rentang usia.
